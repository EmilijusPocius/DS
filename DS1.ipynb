
Download Dataset



```python
import shutil
import kagglehub
import os

path = kagglehub.dataset_download("chik0di/board-games-dataset-complete-features") # # Download (to cache)

target_path = "./boardgame_data"  # Target path (Colab current directory)
if not os.path.exists(target_path):  # Copy extracted dataset here
    shutil.copytree(path, target_path)
    print("Dataset copied to:", os.path.abspath(target_path))
```

Import libraries

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

Load Dataset

```python
df = pandas.read_csv('./boardgame_data/boardgame-geek-dataset_organized.csv')
```

```python
df.info()
df.head()
```

```python
df.describe()
```

```python
nan_list = df.isna().sum().sort_values(ascending=False)
print(nan_list)
```

```python
df = df.dropna(axis='columns');
nan_list = df.isna().sum()
print(nan_list)
```

```python
avg_rating = df['avg_rating']
fig, ax = plt.subplots()

ax.hist(avg_rating,  linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 10), xticks=np.arange(1, 10))

plt.xlabel("Average rating")
plt.ylabel("Amount of games rated")

plt.show()
```

```python
fig, ax = plt.subplots()
colors = np.random.uniform(1, 10, len(df))

ax.scatter(df['num_ratings'], df['avg_rating'], c=colors)
plt.xlabel("Amount of ratings")
plt.ylabel("Average rating")

plt.show()
```

```python
fig, ax = plt.subplots()
colors = np.random.uniform(1, 2, len(df))
ax.scatter(df['complexity'], df['avg_rating'], c=colors)
plt.xlabel("Game complexity")
plt.ylabel("Average rating")

plt.show()
```

```python
fig, ax = plt.subplots()
colors = np.random.uniform(1, 2, len(df))
ax.scatter(df['max_playtime'], df['avg_rating'], c=colors)
plt.xlabel("Maximum playtime")
plt.ylabel("Average rating")

plt.show()
```
